```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# import ssn.preprocessing as encoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Iris.csv')
```

```python
def fun(y):
  return 0 if(y<0) else 0.5 if(y<0.5) else 1
```

```python
x= df[['PetalWidthCm','PetalLengthCm']]
```

```python
encoder = LabelEncoder()
lable= encoder.fit_transform(df.Species)
```

```python
epochs = 50
```

```python
w1 = w2 = b = alpha = 0.25

for epoch in range(epochs):
  mse=0
  print(f"For epoch {epoch +1}:")
  print("========================")
  for i in range(len(x['PetalLengthCm'])):
    print(f"For sample : {i +1} : ")
    x1 , x2 = x['PetalWidthCm'][i], x['PetalLengthCm'][i]
    actual = lable[i]
    yin = round((w1*x1)+(w2*x2)+b,4)
    print(f"Yin = {yin}")
    
    error = actual - fun(yin)
    print(f"Error = {error}")

    mse +=error**2

    w1 = round((w1 + (alpha*(error)*x1)),4)
    w2 = round((w2 + (alpha*(error)*x2)),4)
    b = round((b + (alpha*error)),4)

    print(f"New w1 = { w1 }, new w2 = { w2 }, new b = { b }")
    print()
    print("========================")
  print(f"MSE = {mse}")


```
